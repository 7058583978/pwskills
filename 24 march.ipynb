```python
as we can see here there is no missing values so need to handle missing values
There are several techniques that can be used to handle missing data in a dataset, including:

Mean/median imputation: This involves replacing missing values with the mean or median of the remaining 
values in the same column. This method is simple and easy to implement, but it can lead to biased estimates
of the true values, especially if the missing data is not missing at random.

Mode imputation: This involves replacing missing values with the mode (most frequent value) of the remaining
values in the same column. This method is useful for categorical data, but it can also lead to biased estimates
if the mode is not representative of the true values.

Regression imputation: This involves using regression analysis to predict the missing values based on the values
of other variables in the dataset. This method can provide accurate estimates of the missing values, but it can 
be computationally intensive and may require a large amount of data.

K-nearest neighbor imputation: This involves finding the k nearest neighbors to the observation with the missing 
value and using their values to estimate the missing value. This method can be effective for small datasets, but
it may not work well for large datasets or datasets with high dimensionality.

Advantages and disadvantages of different imputation techniques:

Mean/median imputation:

Advantages: Easy to implement, preserves the mean and variance of the variable.
Disadvantages: May introduce bias in the estimates of the true values, does not take into account the correlation 
between variables.
Mode imputation:

Advantages: Useful for categorical data, easy to implement.
Disadvantages: May introduce bias in the estimates of the true values, does not take into account the correlation
between variables.
Regression imputation:

Advantages: Can provide accurate estimates of missing values, takes into account the correlation between variables.
Disadvantages: Computationally intensive, requires a large amount of data, assumes a linear relationship between 
variables.
K-nearest neighbor imputation:

Advantages: Can be effective for small datasets, takes into account the correlation between variables.
Disadvantages: May not work well for large datasets or datasets with high dimensionality, sensitive to the choice 
of k.
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset = pd.read_csv('winequality-red.csv')
X = dataset.iloc[:, 0:11].values
y = dataset.iloc[:, 11].values
```

```python
from sklearn.model_selection import train_test_split
 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
 
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.decomposition import PCA
 
pca = PCA(n_components = 2)
 
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
 
explained_variance = pca.explained_variance_ratio_
```

```python
from sklearn.linear_model import LogisticRegression 
 
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
from sklearn.metrics import confusion_matrix
 
cm = confusion_matrix(y_test, y_pred)
```

```python
from matplotlib.colors import ListedColormap
 
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1,
                     stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1,
                     stop = X_set[:, 1].max() + 1, step = 0.01))
 
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(),
             X2.ravel()]).T).reshape(X1.shape), alpha = 0.75,
             cmap = ListedColormap(('yellow', 'white', 'aquamarine')))
 
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
 
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
    c = ListedColormap(('r', 'g', 'b'))(i), label = j)
 

plt.xlabel('PC1') 
plt.ylabel('PC2') 
plt.legend() 
 
plt.show()
```
