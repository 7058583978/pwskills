```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
ans] subplot used to plot the multiple figures in single chart so that
     we can compare multiple charts
```

```python
import matplotlib.pyplot as bb
import numpy as np
```

```python
bb.figure(figsize=(10,10))
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
bb.subplot(1,4,1)
bb.plot(x,y)
bb.title("line 1 ")

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
bb.subplot(1,4,2)
bb.plot(x,y)
bb.title("line 2")

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
bb.subplot(1,4,3)
bb.plot(x,y)
bb.title("line 3")

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
bb.subplot(1,4,4)
bb.plot(x,y)
bb.title("line 4")
```

```python
ans] box charts are used to plot the data roaming around the median for that we have to calculate
mean max mode and median.it is very useful chart in case we want to know how a perticular point changes arount it

ex : candle fraph in stocks
```

```python
import matplotlib.pyplot as bb
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
bb.figure(figsize=(10,10))
bb.subplot(1,2,1)
bb.boxplot(box1)

bb.subplot(1,2,2)
bb.boxplot(box2)
```
