```python
Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
```

```python
ans]glyphs are nothing but the the shape of markers used to represent the data 
eg. circles,diamond,squares etc

simply we need to define the marker shape as we want

x = [2,3,5,4,6]
y = [1,2,3,4,6]

output_file("line.html")
p = figure(title = "line plot")
p.scatter(x,y,fill_color = "green",marker = "square",size = 20)
show(p)
```

```python
Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
```

```python
ans]
x = [20,30,50,40,60]
y = [10,2,30,4,60]

output_file("line.html")
p = figure(title = "line plot")
p.xaxis.axis_label = "x axis"
p.yaxis.axis_label = "y axis"
p.scatter(x,y,fill_color = "red",marker = "circle",size = 20,legend="red points")
show(p)

so in above data we have introduced the title,axes,and legend which is used to decorate our graph and make our figure
more detailed
```
